```python
new_data=pd.read_csv(r'C:\Users\DELL\Downloads\internship-test\internship-test\filteredCountry.csv')
```

```python
new_data_df = pd.DataFrame(new_data, columns= ['SKU','PRICE'])
```

```python
new_data_df.head()
```

```python
new_data_df['PRICE'] = new_data_df['PRICE'].str.replace('$', '')
new_data_df['PRICE'] = new_data_df['PRICE'].str.replace(',', '.')
new_data_df['PRICE'] = new_data_df['PRICE'].str.replace('?', '')
```

```python
new_data_df.head(10)
```

```python
new_data_df["PRICE"] = pd.to_numeric(new_data_df["PRICE"])
```

```python
SKU_data=new_data_df.groupby("SKU")
SKU_data.head()
```

```python
#SKU_data["PRICE"] = pd.to_numeric(SKU_data["PRICE"])
```

```python
result = new_data_df.groupby('SKU').agg({'PRICE': ['min','max']})
result.head(10)
```

```python
feilds=["FIRST_MINIMUM_PRICE","SECOND_MINIMUM_PRICE"]
```

```python
result.columns = ["".join(x) for x in feilds]
```
